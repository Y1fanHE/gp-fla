```python
import matplotlib.pyplot as plt
from plot_tools import get_evo_curve
plt.style.use("bmh")
```

## Parity

```python
problems = ["parity4", "parity6", "parity9", "parity11"]
titles = ["Parity 4", "Parity 6", "Parity 9", "Parity 11"]
fig, ax = plt.subplots(1, 4, figsize=(15,3))
plt.subplots_adjust(wspace=0.15)
for i, p in enumerate(problems):
    map_ht = get_evo_curve(f"dat/{p}-gp-map.zip")
    map_hd = get_evo_curve(f"dat/{p}-gp-map1.zip")
    map_hd20 = get_evo_curve(f"dat/{p}-gp-map2.zip")
    map_hb = get_evo_curve(f"dat/{p}-gp-map3.zip")
    plain = get_evo_curve(f"dat/{p}-gp.zip")
    l1 = ax[i].plot(range(1, len(map_ht)+1),   map_ht,   label="MAP-HT",   zorder=6)
    l2 = ax[i].plot(range(1, len(map_hd)+1),   map_hd,   label="MAP-HD",   zorder=5, ls="--")
    l3 = ax[i].plot(range(1, len(map_hd20)+1), map_hd20, label="MAP-HD20", zorder=4, ls="--")
    l4 = ax[i].plot(range(1, len(map_hb)+1),   map_hb,   label="MAP-HB",   zorder=3, ls="-.")
    l5 = ax[i].plot(range(1, len(plain)+1),    plain,    label="Plain",    zorder=2, c="black", ls=":")
    ax[i].set_xlabel("Generation")
    if i==0:
        ax[i].set_ylabel("Success Rate")
    ax[i].set_xscale("log")
    ax[i].set_ylim(-0.03, 1.03)
    ax[i].spines[["right","top"]].set_visible(False)
    ax[i].set_title(titles[i])
fig.legend([l1,l2,l3,l4,l5],
           labels=["MAP-HT", "MAP-HD", "MAP-HD20", "MAP-HB", "Plain"],
           ncol=5,
           frameon=False,
           loc="lower center",
           bbox_to_anchor=(0.5,-0.2))
plt.savefig(f"evo/parity.png", dpi=300, bbox_inches="tight")
plt.show()
```
