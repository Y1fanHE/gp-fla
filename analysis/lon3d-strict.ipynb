```python
from plot_tools import get_networkx_plot_var, set_ax
import matplotlib.pyplot as plt
```

```python
vertex_xyz, edge_xyz, vertex_size, edge_width, vertex_color, edge_color = get_networkx_plot_var("dat/parity4-5.zip", 0, 16, monotonic=True, strict=True)
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection="3d")
for i, vizedge in enumerate(edge_xyz):
    ax.plot(*vizedge.T, color=edge_color[i], lw=edge_width[i])
ax.scatter(*vertex_xyz.T, s=vertex_size, ec="black", fc=vertex_color, alpha=1.)
set_ax(ax)
plt.savefig("lons/parity4-5-3d-strict.svg", bbox_inches="tight")
```
