```python
from zipfile import ZipFile
import pandas as pd
import numpy as np
from scipy.stats import entropy
```

```python
with open("neu/dat.csv", "w") as f:
    f.close()
for zfname in [
    "parity4-5", "parity4-7", "parity4-9",
    "parity6-5", "parity6-7", "parity6-9",
    "ant1-5", "ant1-7", "ant1-9"
]:
    zf = ZipFile(f"dat/{zfname}-rw.zip")
    eps = 0
    neu = 0
    rug = 0
    for i in zf.namelist():
        fit = pd.read_csv(zf.open(i), header=None, delimiter=" ")[0].values
        diff = fit[:-1] - fit[1:]
        neu += sum(abs(diff) <= eps) / len(diff)

        diff_ = np.full(diff.shape, 0)
        diff_[diff<-eps] = -1
        diff_[diff>eps] = 1
        tmp = np.vstack([diff_[:-1], diff_[1:]]).T
        tmp = tmp[np.where(tmp[:,0]!=tmp[:,1])]
        tmp = tmp[:,0]*10+tmp[:,1]
        prob = pd.Series(tmp).value_counts()
        rug += entropy(prob, base=6)

    with open("neu/dat.csv", "a") as f:
        print(zfname, eps, neu/len(zf.namelist()), rug/len(zf.namelist()), sep=",", file=f)
```
