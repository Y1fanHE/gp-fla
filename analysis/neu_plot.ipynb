```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("bmh")
```

```python
dat = pd.read_csv("neu/dat.csv", header=None, delimiter=",")
dat.columns = ["instance", "eps", "neu", "rug"]
```

```python
problems = ["nguyen2", "nguyen9"]
titles = ["Nguyen 2", "Nguyen 9"]
fig, ax = plt.subplots(1, 4, figsize=(15,3))
plt.subplots_adjust(wspace=0.15)
for i, p in enumerate(problems):
    dat_ = dat[dat.instance==f"{p}-5"]
    l1 = ax[i+1].plot(dat_.eps, dat_.rug, ls="-", label="Max Height=5")
    dat_ = dat[dat.instance==f"{p}-7"]
    l2 = ax[i+1].plot(dat_.eps, dat_.rug, ls="--", label="Max Height=7")
    dat_ = dat[dat.instance==f"{p}-9"]
    l3 = ax[i+1].plot(dat_.eps, dat_.rug, ls=":", label="Max Height=9")
    ax[i+1].set_xlabel("Threshold")
    if i==0:
        ax[i+1].set_ylabel("Ruggedness")
    ax[i+1].set_ylim(-0.03, 1.03)
    ax[i+1].set_title(titles[i])
ax[0].set_visible(False)
ax[3].set_visible(False)
fig.legend([l1,l2,l3],
           labels=[f"Max Height={h}" for h in [5,7,9]],
           ncol=3,
           frameon=False,
           loc="lower center",
           bbox_to_anchor=(0.5,-0.2))
plt.savefig("neu/nguyen-rug.svg", bbox_inches="tight")
```
